Нужно создать систему, которая могла бы оценить риск ДТП по выбранному маршруту движения. Под риском понимается вероятность ДТП с любым повреждением транспортного средства. Как только водитель забронировал автомобиль, сел за руль и выбрал маршрут, система должна оценить уровень риска. Если уровень риска высок, водитель увидит предупреждение и рекомендации по маршруту.

Идея создания такой системы находится в стадии предварительного обсуждения и проработки. Чёткого алгоритма работы и подобных решений на рынке ещё не существует. Текущая задача — понять, возможно ли предсказывать ДТП, опираясь на исторические данные одного из регионов.

**Идея решения задачи от заказчика:**
1. Создать модель предсказания ДТП (целевое значение — `at_fault` (виновник) в таблице `parties`)
    - Для модели выбрать тип виновника — только машина (`car`).
    - Выбрать случаи, когда ДТП привело к любым повреждениям транспортного средства, кроме типа `SCRATCH` (царапина).
    - Для моделирования ограничиться данными за 2012 год — они самые свежие.
    - Обязательное условие — учесть фактор возраста автомобиля.
2. На основе модели исследовать основные факторы ДТП.
3. Понять, помогут ли результаты моделирования и анализ важности факторов ответить на вопросы:
    - Возможно ли создать адекватную системы оценки водительского риска при выдаче авто?
    - Какие ещё факторы нужно учесть?
    - Нужно ли оборудовать автомобиль какими-либо датчиками или камерой?
    
Заказчик предлагает поработать с базой данных по происшествиям и сформировать идеи создания такой системы.

```python
def table_info(table_name, engine):
    # Создаем инспектор для получения метаданных
    inspector = inspect(engine)
    
    # Получаем информацию о столбцах таблицы
    columns = inspector.get_columns(table_name)
    print(f"Структура таблицы {table_name}:")
    for column in columns:
        print(f"{column['name']}: {column['type']} {'(NULL)' if column['nullable'] else '(NOT NULL)'}")
    
    # Получаем количество строк в таблице
    row_count = engine.execute(f"SELECT COUNT(*) FROM {table_name}").scalar()
    print(f"\nКоличество строк в таблице: {row_count}")
    
    # Получаем первые 5 строк
    print(f"\nПервые пять строк таблицы {table_name}:")
    sample_data = pd.read_sql(f"SELECT * FROM {table_name} LIMIT 5", con=engine)
    display(sample_data)
    
    # Анализ уникальных значений с добавлением статистик
    print("\nАнализ уникальных значений и статистик:")
    unique_stats = []
    for column in columns:
        col_name = column['name']
        # Количество уникальных значений
        unique_count = engine.execute(f"SELECT COUNT(DISTINCT {col_name}) FROM {table_name}").scalar()
        # Процент уникальных значений
        unique_percent = (unique_count / row_count) * 100 if row_count > 0 else 0
        
        # Для числовых столбцов получаем статистики
        if hasattr(column['type'], 'python_type') and issubclass(column['type'].python_type, (int, float)):
            stats_query = f"""
            SELECT 
                AVG({col_name}) as mean_val,
                STDDEV({col_name}) as std_val,
                MIN({col_name}) as min_val,
                PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY {col_name}) as median_val,
                MAX({col_name}) as max_val
            FROM {table_name}
            WHERE {col_name} IS NOT NULL
            """
            stats_result = engine.execute(stats_query).fetchone()
        else:
            stats_result = None
        
        # Для нечисловых столбцов получаем самое частое значение и его частоту
        if hasattr(column['type'], 'python_type') and not issubclass(column['type'].python_type, (int, float)):
            top_query = f"""
            SELECT {col_name} as top_value, COUNT(*) as frequency 
            FROM {table_name} 
            WHERE {col_name} IS NOT NULL 
            GROUP BY {col_name} 
            ORDER BY frequency DESC 
            LIMIT 1
            """
            top_result = engine.execute(top_query).fetchone()
            top_value = top_result[0] if top_result else None
            frequency = top_result[1] if top_result else 0
            freq_percent = (frequency / row_count) * 100 if row_count > 0 else 0
        else:
            top_value = None
            frequency = None
            freq_percent = None
        
        unique_stats.append({
            'Column': col_name,
            'Unique Values': unique_count,
            'Unique %': f"{unique_percent:.2f}%",
            'Most Frequent Value': top_value,
            'Frequency': frequency,
            'Frequency %': f"{freq_percent:.2f}%" if freq_percent is not None else None,
            'Mean': stats_result[0] if stats_result else None,
            'Std': stats_result[1] if stats_result else None,
            'Min': stats_result[2] if stats_result else None,
            'Median (50%)': stats_result[3] if stats_result else None,
            'Max': stats_result[4] if stats_result else None
        })
    
    unique_stats_df = pd.DataFrame(unique_stats)
    display(unique_stats_df)
    
    # Получаем информацию о пропущенных значениях
    print("\nКоличество пропущенных значений по столбцам:")
    null_counts = {}
    for column in columns:
        col_name = column['name']
        null_count = engine.execute(f"SELECT COUNT(*) FROM {table_name} WHERE {col_name} IS NULL").scalar()
        null_counts[col_name] = null_count
    
    null_counts_df = pd.DataFrame.from_dict(null_counts, orient='index', columns=['Null Count'])
    null_counts_df['Null %'] = (null_counts_df['Null Count'] / row_count * 100).round(2).astype(str) + '%'
    display(null_counts_df)
    
    return sample_data
```

### Подготовка набора данных

Подготовим набор данных на основе первичного предположения заказчика: 
- Тип виновника — только машина (`car`).
- Случаи, когда ДТП привело к любым значимым повреждениям автомобиля любого из участников — все, кроме типа `SCRATCH` (царапина).
- Данные только за 2012 год.

```python
query ='''
SELECT   p.at_fault,
         p.cellphone_in_use,
         c.distance,
         c.direction,
         c.weather_1 AS weather,
         c.location_type,
         c.road_surface,
         c.road_condition_1 AS road_condition,
         c.lighting,
         c.control_device,
         TO_CHAR(c.collision_date, 'FMDay') as day_of_week,
         TO_CHAR(c.collision_time, 'HH24') as hour,
         v.vehicle_type,
         v.vehicle_transmission,
         v.vehicle_age
FROM parties AS P
JOIN vehicles AS v ON p.case_id=v.case_id AND p.party_number=v.party_number
JOIN collisions AS c ON p.case_id=c.case_id 
WHERE p.party_type = 'car' AND c.collision_damage != 'scratch' AND EXTRACT(YEAR FROM c.collision_date) = 2012; 
'''
data = pd.read_sql_query(query, con=engine) 
```

```python
def convert_dtypes(data):
    # Категориальные признаки
    cat_cols = [
        'at_fault', 'cellphone_in_use', 'direction', 'weather', 
        'location_type', 'road_surface',
        'road_condition', 'lighting', 'control_device',
        'vehicle_type', 'vehicle_transmission', 'day_of_week'
    ]
    
    # Числовые признаки
    num_cols = {
        'distance': 'float',
        'hour': 'int64',
        'vehicle_age': 'int64'
    }
    
    # 1. Обработка числовых колонок
    for col, dtype in num_cols.items():
        if col in data.columns:
            data[col] = pd.to_numeric(data[col], errors='coerce').astype(dtype)
    
    # 2. Обработка категориальных колонок
    for col in cat_cols:
        if col in data.columns:
            data[col] = data[col].astype('category')
    
    # 3. Нормализация cellphone_in_use
    if 'cellphone_in_use' in data.columns:
        data['cellphone_in_use'] = data['cellphone_in_use'].replace({
            0.0: 'no',
            1.0: 'yes',
            'unknown': 'unknown'
        }).astype('category')
    
    return data

data = convert_dtypes(data)
```

```python
numeric_features = ['vehicle_age']
categorical_features = X.select_dtypes(include=['category']).columns.tolist()

preprocessor_ohe = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features), 
        ('cat', OneHotEncoder(handle_unknown='ignore', drop='first'), categorical_features)
    ])
```

```python
preprocessor_oe = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features), 
        ('cat', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), categorical_features)
    ])
```
